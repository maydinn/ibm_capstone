```python
df_ = pd.melt(df, value_vars=list(df.columns))
df_.drop('variable', axis=1, inplace = True)
df_
```

```python
df_['postal_code']= df_['value'].apply(lambda x: x[:3])
```

```python
def boroug(x): 
    a = ""
    for i in x[3:]:
        if i != "(":
            a += i
        else:
            break
    return a
```

```python
df_['boroug']=df_['value'].apply(boroug)
df_.sample(6)
```

```python
def get(x):
    re = ""
    flag = False
    for i in x:
        if i == '(':
            flag = True
            continue
        elif i == ')':
            flag = False
        if flag:
            re += i
    return re
```

```python
ssd = []
X = df_l[['Latitude','Longitude']]

K = range(1,10)

for k in K:
    model = KMeans(n_clusters=k)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(K, ssd, "bx-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error")
plt.title("Elbow Method")
```

```python
from yellowbrick.cluster import KElbowVisualizer
kmeans = KMeans()
visu = KElbowVisualizer(kmeans, k = (2,10))
visu.fit(X)
visu.poof()
```
