```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
df_ = pd.melt(df, value_vars=list(df.columns))
df_.drop('variable', axis=1, inplace = True)
df_
```

```python
df_['postal_code']= df_['value'].apply(lambda x: x[:3])
```

```python
df_
```

```python
def boroug(x): 
    a = ""
    for i in x[3:]:
        if i != "(":
            a += i
        else:
            break
    return a
```

```python
df_['boroug']=df_['value'].apply(boroug)
df_.sample(6)
```

```python
df_.sample(6)
```

```python
df_.boroug=df_.boroug.apply(lambda x: x if x==x.title() else np.NaN)
df_
```

```python
df_.dropna(inplace=True)
```

```python
df_
```

```python
def get(x):
    re = ""
    flag = False
    for i in x:
        if i == '(':
            flag = True
            continue
        elif i == ')':
            flag = False
        if flag:
            re += i
    return re
```

```python
get('M1BScarborough(Malvern / Rouge)')
```

```python
df_['near'] = df_.value.apply(get)
```
